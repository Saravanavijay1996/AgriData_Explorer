```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data Read
data = pd.read_csv("Agri_Data.csv")
data
```

```python
# duplicates
duplicates = data[data.duplicated()]
print("total duplicates rows:",len(duplicates))
```

```python
# Delete Unwanted Columns

data = data.drop(["Dist Code","State Code","BARLEY AREA (1000 ha)","BARLEY PRODUCTION (1000 tons)","BARLEY YIELD (Kg per ha)","CHICKPEA AREA (1000 ha)","CHICKPEA PRODUCTION (1000 tons)","CHICKPEA YIELD (Kg per ha)","PIGEONPEA AREA (1000 ha)","PIGEONPEA PRODUCTION (1000 tons)","PIGEONPEA YIELD (Kg per ha)","MINOR PULSES AREA (1000 ha)","MINOR PULSES PRODUCTION (1000 tons)","MINOR PULSES YIELD (Kg per ha)","FRUITS AREA (1000 ha)","VEGETABLES AREA (1000 ha)","FRUITS AND VEGETABLES AREA (1000 ha)","POTATOES AREA (1000 ha)","ONION AREA (1000 ha)","FODDER AREA (1000 ha)","SESAMUM AREA (1000 ha)","SESAMUM PRODUCTION (1000 tons)","SESAMUM YIELD (Kg per ha)","RAPESEED AND MUSTARD AREA (1000 ha)","RAPESEED AND MUSTARD PRODUCTION (1000 tons)","RAPESEED AND MUSTARD YIELD (Kg per ha)","SAFFLOWER AREA (1000 ha)","SAFFLOWER PRODUCTION (1000 tons)","SAFFLOWER YIELD (Kg per ha)","CASTOR AREA (1000 ha)","CASTOR PRODUCTION (1000 tons)","CASTOR YIELD (Kg per ha)","LINSEED AREA (1000 ha)","LINSEED PRODUCTION (1000 tons)","LINSEED YIELD (Kg per ha)"], axis=1)
data
```

```python
# 1st Question

# Top 7 RICE PRODUCTION State Data(Bar_plot)
data_rice = data[["State Name","Year","RICE PRODUCTION (1000 tons)"]]

state_rice = data_rice.groupby("State Name")["RICE PRODUCTION (1000 tons)"].sum().reset_index()

top7_rice_states = state_rice.sort_values(by="RICE PRODUCTION (1000 tons)", ascending=False).head(7)
```

```python
# Bar Plot
plt.figure(figsize=(12, 6))

sns.barplot(data=top7_rice_states, x="RICE PRODUCTION (1000 tons)", y ="State Name" , palette= "viridis")

plt.title("Top 7 Rice Producing States in India",fontsize=14)

plt.xlabel("Total Rice Production(1000 tons)")

plt.ylabel("State")

plt.show()
```

```python
# 2nd Question
# Top 5 Wheat Producing States Data(Bar_chart)and its percentage(%)(Pie_chart) 
data_wheat = data[["State Name","Year","WHEAT PRODUCTION (1000 tons)"]]
state_wheat = data_wheat.groupby("State Name")["WHEAT PRODUCTION (1000 tons)"].sum().reset_index()
top5_wheat_states = state_wheat.sort_values(by="WHEAT PRODUCTION (1000 tons)", ascending=False).head(5)
top5_wheat_states["Percentage"] = (top5_wheat_states["WHEAT PRODUCTION (1000 tons)"]/top5_wheat_states["WHEAT PRODUCTION (1000 tons)"].sum())*100
```

```python
# Bar plot
plt.figure(figsize=(12, 6))

sns.barplot(data=top5_wheat_states, x="WHEAT PRODUCTION (1000 tons)", y ="State Name" , color= "Green")

plt.title("Top 5 Wheat Producing States in India",fontsize=14)

plt.xlabel("Total Wheat Production(1000 tons)")

plt.ylabel("State")

plt.show()
```

```python
# Pie Chart
plt.figure(figsize=(5,5))
plt.pie(
   top5_wheat_states["Percentage"],
    labels = top5_wheat_states["State Name"],
    autopct="%.1f%%",
    startangle = 140
)
plt.title("Percentage Share of Top 5 Wheat Producing States",fontsize=12)

plt.show()
```

```python
# 3rd Question
# Oil seed production by top 5 states
data_oil = data[["State Name","Year","OILSEEDS PRODUCTION (1000 tons)"]]
state_oil = data_oil.groupby("State Name")["OILSEEDS PRODUCTION (1000 tons)"].sum().reset_index()
top5_oil_states = state_oil.sort_values(by="OILSEEDS PRODUCTION (1000 tons)", ascending=False).head(5)
```

```python
# Line
plt.figure(figsize=(10, 6))

sns.lineplot(data=top5_oil_states, x="OILSEEDS PRODUCTION (1000 tons)", y ="State Name" , color= "Brown", marker = "o")

plt.title("Top 5 Oilseeds Producing States in India",fontsize=14)

plt.xlabel("Total Oilseeds Production(1000 tons)")

plt.ylabel("State")

plt.legend()

plt.show()
```

```python
# 4th Question
# Top 7 SUNFLOWER PRODUCTION  State
data_sunflower = data[["State Name","Year","SUNFLOWER PRODUCTION (1000 tons)"]]
state_sunflower = data_sunflower.groupby("State Name")["SUNFLOWER PRODUCTION (1000 tons)"].sum().reset_index()
top7_sunflower_states = state_sunflower.sort_values(by="SUNFLOWER PRODUCTION (1000 tons)", ascending=False).head(7)
```

```python
# Bar Plot
colors = ["pink","green","yellow","brown","orange","grey","gold"]

plt.figure(figsize=(10, 6))

plt.bar(top7_sunflower_states["State Name"],top7_sunflower_states["SUNFLOWER PRODUCTION (1000 tons)"], color=colors)

plt.title("Top 7 Sunflower Producing States in India",fontsize=12)

plt.xlabel("Total Sunflower Production(1000 tons)")

plt.ylabel("State")

plt.show()
```

```python
# 5th Question
# India's SUGARCANE PRODUCTION From Last 50 Years(Line_plot)
print("Earliest year:",data["Year"].min())
print("Latest year:",data["Year"].max())
```

```python
data_sugarcane = data[["State Name","Year","SUGARCANE PRODUCTION (1000 tons)"]]
india_sugarcane = data_sugarcane.groupby("Year")["SUGARCANE PRODUCTION (1000 tons)"].sum().reset_index()
india_sugarcane_last50 = india_sugarcane.tail(50)
```

```python
# Line Plot
plt.figure(figsize=(12, 6))

plt.plot(india_sugarcane_last50["Year"],india_sugarcane_last50["SUGARCANE PRODUCTION (1000 tons)"], marker="o", color="lightblue", linewidth=2)
plt.title("India's Sugarcane Production (Last 50 Years)",fontsize=14)

plt.xlabel("Year")

plt.ylabel("Sugarcane Production (1000 tons)")

plt.grid(True,linestyle="--",alpha=0.6)

plt.show()
```

```python
# 6th Question
# Rice Production Vs Wheat Production (Last 50y)
data_ricewheat = data[["Year","RICE PRODUCTION (1000 tons)","WHEAT PRODUCTION (1000 tons)"]]
india_ricewheat = data.groupby("Year")[["RICE PRODUCTION (1000 tons)","WHEAT PRODUCTION (1000 tons)"]].sum().reset_index()
india_ricewheat_last50 = india_ricewheat.sort_values("Year").tail(50)
```

```python
# Stacked Area Chart
plt.figure(figsize=(12, 6))

plt.plot(india_ricewheat_last50["Year"],india_ricewheat_last50["RICE PRODUCTION (1000 tons)"],marker="o",color="green",linewidth=2,label="Rice")

plt.plot(india_ricewheat_last50["Year"],india_ricewheat_last50["WHEAT PRODUCTION (1000 tons)"],marker="o",color="red",linewidth=2,label="Wheat")

plt.xlabel("Year")
plt.ylabel("Production (1000 tons)")
plt.title("Rice vs Wheat Production in India (Last 50 Years)")
plt.legend()
plt.grid(True,linestyle="--",alpha=0.6)
plt.show()
```

```python
# 7th Question
# Rice Production By West Bengal Districts
data_westbengal = data[(data["State Name"]=="West Bengal")][["Dist Name","Year","RICE PRODUCTION (1000 tons)"]]
westbengal_rice = data_westbengal.groupby("Dist Name")["RICE PRODUCTION (1000 tons)"].sum().reset_index()
westbengal_rice = westbengal_rice.sort_values("RICE PRODUCTION (1000 tons)",ascending=False)
```

```python
# Donut Chart
plt.figure(figsize=(5,5))
plt.pie(
    westbengal_rice["RICE PRODUCTION (1000 tons)"],
    labels=westbengal_rice["Dist Name"],
    autopct="%1.1f%%",
    startangle=140,
    pctdistance=0.85,
    colors=plt.cm.Paired.colors  # attractive palette
)

# Draw circle in middle (to make donut)
centre_circle = plt.Circle((0,0), 0.70, fc="white")
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("District-wise Rice Production in West Bengal (Donut Chart)")
plt.tight_layout()
plt.show()
```

```python
# 8th Question
# Top 10 Wheat Production Years From UP
data_uttarpradesh = data[data["State Name"] == "Uttar Pradesh"][["Year", "WHEAT PRODUCTION (1000 tons)"]]
uttarpradesh_wheat = data_uttarpradesh.groupby("Year")["WHEAT PRODUCTION (1000 tons)"].sum().reset_index()
top10_wheat = uttarpradesh_wheat.sort_values("WHEAT PRODUCTION (1000 tons)", ascending=False).head(10)
```

```python
# Bar Plot

colors = ["red", "yellow", "orange", "green", "lightblue", 
          "violet", "blue", "grey", "brown", "pink"]
plt.figure(figsize=(10,6))
bars = plt.bar(top10_wheat["Year"].astype(str), 
        top10_wheat["WHEAT PRODUCTION (1000 tons)"], 
        color=colors[:len(top10_wheat)])

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, 
             f"{int(height):,}",  # formatted with commas
             ha="center", va="bottom", fontsize=9, fontweight="bold")

plt.xlabel("Year")
plt.ylabel("Wheat Production (1000 tons)")
plt.title("Top 10 Wheat Production Years - Uttar Pradesh")
plt.xticks(rotation=45)
plt.show()
```

```python
# 9th Question
# Millet Production (Last 50y)
data_millets = data.groupby("Year")[[
    "SORGHUM PRODUCTION (1000 tons)",
    "RABI SORGHUM PRODUCTION (1000 tons)",
    "KHARIF SORGHUM PRODUCTION (1000 tons)",
    "PEARL MILLET PRODUCTION (1000 tons)",
    "FINGER MILLET PRODUCTION (1000 tons)"
]].sum().reset_index()
data_millets_last50 = data_millets.sort_values("Year").tail(50)
```

```python
# Stacked Bar Chart
data_millets_last50.set_index("Year")[[
    "SORGHUM PRODUCTION (1000 tons)",
    "RABI SORGHUM PRODUCTION (1000 tons)",
    "KHARIF SORGHUM PRODUCTION (1000 tons)",
    "PEARL MILLET PRODUCTION (1000 tons)",
    "FINGER MILLET PRODUCTION (1000 tons)"
]].plot(kind="bar", stacked=True, figsize=(14,6), colormap="tab20")

plt.title("Millet Production in India (Stacked Bar - Last 50 Years)")
plt.ylabel("Production (1000 tons)")
plt.xlabel("Year")
plt.legend(title="Millet Type")
plt.show()
```

```python
# 10th Question
# Sorghum Production (Kharif and Rabi) by Region 
data_sorghum_region = data.groupby("State Name")[[
    "KHARIF SORGHUM PRODUCTION (1000 tons)",
    "RABI SORGHUM PRODUCTION (1000 tons)"
]].sum().reset_index()
```

```python
# 11th Question
# Top 7 States for Groundnut Production
data_groundnut = data.groupby("State Name")[["GROUNDNUT PRODUCTION (1000 tons)"]].sum().reset_index()
top7_groundnut = data_groundnut.sort_values("GROUNDNUT PRODUCTION (1000 tons)", ascending=False).head(7)
```

```python
# Bar Chart
plt.figure(figsize=(10,6))
colors = ["orange","blue","red","green","violet","grey","lightblue"]

plt.bar(top7_groundnut["State Name"], top7_groundnut["GROUNDNUT PRODUCTION (1000 tons)"], color=colors)

plt.xlabel("State")
plt.ylabel("Groundnut Production (1000 tons)")
plt.title("Top 7 Groundnut Producing States in India")
plt.xticks(rotation=30, ha="right")
plt.show()
```

```python
# 12th Question
# Soyabean Production by Top 5 States and Yield Efficiency
data_soyabean = data.groupby("State Name")[[
    "SOYABEAN PRODUCTION (1000 tons)",
    "SOYABEAN AREA (1000 ha)"
]].sum().reset_index()

# Calculate yield efficiency
data_soyabean["Yield Efficiency (tons/ha)"] = data_soyabean["SOYABEAN PRODUCTION (1000 tons)"] / data_soyabean["SOYABEAN AREA (1000 ha)"]

top5_soyabean = data_soyabean.sort_values("SOYABEAN PRODUCTION (1000 tons)", ascending=False).head(5)
```

```python
# Line Chart for Soyabean Production
# Line Plot
plt.figure(figsize=(10, 6))

sns.lineplot(data=top5_soyabean, x="SOYABEAN PRODUCTION (1000 tons)", y ="State Name" , color= "skyblue", marker = "o")

plt.title("Top 5 Soyabean Producing States in India",fontsize=14)

plt.xlabel("Total Soyabean Production(1000 tons)")

plt.ylabel("State Name")

plt.legend()

plt.show()
```

```python
# Bar Chart for Yield Efficiency
x = np.arange(len(top5_soyabean["State Name"]))
width = 0.35  

plt.figure(figsize=(10,6))

plt.bar(x + width/2, top5_soyabean["Yield Efficiency (tons/ha)"], width, label="Yield Efficiency", color="orange")

plt.xticks(x, top5_soyabean["State Name"], rotation=30, ha="right")
plt.ylabel("Value")
plt.title("Soyabean Yield Efficiency (Top 5 States)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 13th Question
#  Oilseed Production in Major States
data_oilseed = data.groupby("State Name")[["OILSEEDS PRODUCTION (1000 tons)"]].sum().reset_index()
data_oilseed_major = data_oilseed.sort_values("OILSEEDS PRODUCTION (1000 tons)", ascending=False).head(10)
```

```python
# 14th Question
# Impact of Area Cultivated on Production (Rice, Wheat, Maize)
data_area_production = data.groupby("Year")[[
    "RICE PRODUCTION (1000 tons)", "RICE AREA (1000 ha)",
    "WHEAT PRODUCTION (1000 tons)", "WHEAT AREA (1000 ha)",
    "MAIZE PRODUCTION (1000 tons)", "MAIZE AREA (1000 ha)"
]].sum().reset_index()
```

```python
# Regression Line Plot
plt.figure(figsize=(12,6))
sns.regplot(x="RICE AREA (1000 ha)", y="RICE PRODUCTION (1000 tons)", data=data_area_production, scatter_kws={'s':40}, label="Rice", color="orange")
sns.regplot(x="WHEAT AREA (1000 ha)", y="WHEAT PRODUCTION (1000 tons)", data=data_area_production, scatter_kws={'s':40}, label="Wheat", color="brown")
sns.regplot(x="MAIZE AREA (1000 ha)", y="MAIZE PRODUCTION (1000 tons)", data=data_area_production, scatter_kws={'s':40}, label="Maize", color="seagreen")

plt.xlabel("Area Cultivated (1000 ha)")
plt.ylabel("Production (1000 tons)")
plt.title("Correlation between Area and Production")
plt.legend()
plt.show()
```

```python
# 15th Question
# Rice vs. Wheat Yield Across States 
data_yield = data.groupby("State Name")[[
    "RICE PRODUCTION (1000 tons)", "RICE AREA (1000 ha)",
    "WHEAT PRODUCTION (1000 tons)", "WHEAT AREA (1000 ha)"
]].sum().reset_index()


data_yield["Rice Yield"] = data_yield["RICE PRODUCTION (1000 tons)"] / data_yield["RICE AREA (1000 ha)"]
data_yield["Wheat Yield"] = data_yield["WHEAT PRODUCTION (1000 tons)"] / data_yield["WHEAT AREA (1000 ha)"]

data_yield = data_yield[["State Name", "Rice Yield", "Wheat Yield"]]
```

```python
# Heatmap Plot
data_heat = data_yield.set_index("State Name")[["Rice Yield","Wheat Yield"]]

plt.figure(figsize=(8,10))
sns.heatmap(data_heat, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Rice vs Wheat Yield Across States (Heatmap)")
plt.show()
```
